```python
import pandas as pd
import numpy as np
import csv
```

```python
#function to open/read the file and create dataframe and prints information of DF
def get_data(file):
    data = pd.read_csv(file)
    df = pd.DataFrame(data.values, columns=data.columns, index=data.index)
    return df
```

```python
#load file and run function
file = 'eur.pop.rev.csv'
EurPopDF = get_data(file)
EurPopDF
```

```python
#Rename columns with dates
EurPopDF = EurPopDF.rename(columns={'European Population (1989-1995)':'Country','Unnamed: 1': '1989','Unnamed: 2': '1990','Unnamed: 3': '1991',
                                   'Unnamed: 4': '1992','Unnamed: 5': '1993','Unnamed: 6': '1994',
                                    'Unnamed: 7': '1995'})
```

```python
#drop first couple of blank columns
EurPopDF = EurPopDF.drop([EurPopDF.index[0], EurPopDF.index[1]])
```

```python
#resetting index
EurPopDF = EurPopDF.reset_index(drop = True)
EurPopDF
```

```python
EurPopDF.info()
```

```python
#create function to replace NAs with 0
def NA_replace(df):
    df = df.fillna(0)
    return df
```

```python
#run function to replace NAs
EurPopDF = NA_replace(EurPopDF)
```

```python
#Changing data type of populations from object to int
EurPopDF[['1989','1990','1991', '1992','1993', '1994','1995']] = EurPopDF[['1989','1990','1991', '1992','1993', '1994','1995']].astype(int)
```

```python
EurPopDF.info()
```

```python
#Print each country that has more than 1,000,000 people in 1995.
print('Countries with over 1,000,000 population in 1995: ')
print(EurPopDF.loc[EurPopDF['1995'] >= 1000000, 'Country'])
```

```python
#Print the average population in the United Kingdom over those seven years.
#First, calculate mean of all rows and adding it to new row at the end:
EurPopDF['mean'] = EurPopDF.mean(axis=1).astype(int)
#saving UK avg to make things easier
UKavg = (EurPopDF.query("Country=='United Kingdom'")["mean"])

#Next, printing the mean of the UK from the years:
print('Average Population of the United Kingdom from 1989 - 1995:')
print('Row #:', UKavg)
```
